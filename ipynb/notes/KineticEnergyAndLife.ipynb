$$
\textbf{1. Postulate momentum and work--energy relation}
$$

$$
p = \Gamma\, m v,
\qquad
\Gamma = \gamma_{\text{life}}\,\gamma_{\text{velocity}}.
$$

$$
dK = F\,dx, 
\quad 
F = \frac{dp}{dt}, 
\quad 
v = \frac{dx}{dt}
\;\Rightarrow\;
dK = v\,dp.
$$

$$
dp = m\big(\Gamma\,dv + v\,d\Gamma\big)
\;\Rightarrow\;
dK = v\,dp = m\big(\Gamma v\,dv + v^2 d\Gamma\big).
$$

$$
\boxed{dK = m\big(\Gamma v\,dv + v^2 d\Gamma\big)}
$$

$$
\textbf{2. Split }\Gamma\text{ into life and velocity factors}
$$

$$
\Gamma = \gamma_{\text{life}}\,\gamma_{\text{velocity}},
\qquad
\gamma_{\text{life}} = \text{constant (along this acceleration)}.
$$

$$
d\Gamma = \gamma_{\text{life}}\,d\gamma_{\text{velocity}}.
$$

$$
dK 
= m\Big(\gamma_{\text{life}}\gamma_{\text{velocity}} v\,dv 
  + v^2 \gamma_{\text{life}}\,d\gamma_{\text{velocity}}\Big)
= m\gamma_{\text{life}}\Big(\gamma_{\text{velocity}} v\,dv 
  + v^2 d\gamma_{\text{velocity}}\Big).
$$

$$
\boxed{
dK 
= m\gamma_{\text{life}}\Big(\gamma_{\text{velocity}} v\,dv 
  + v^2 d\gamma_{\text{velocity}}\Big)
}
$$

$$
\textbf{3. Specify the purely geometric velocity factor}
$$

$$
\gamma_{\text{velocity}} 
= \frac{1}{\sqrt{1 - \dfrac{v^2}{c^2}}}.
$$

Write it as a power:

$$
\gamma_{\text{velocity}} 
= \left(1 - \frac{v^2}{c^2}\right)^{-1/2}.
$$

Differentiate with respect to \(v\):

$$
\frac{d\gamma_{\text{velocity}}}{dv}
= -\frac{1}{2}
\left(1 - \frac{v^2}{c^2}\right)^{-3/2}
\left(-\frac{2v}{c^2}\right)
= \frac{v}{c^2}
\left(1 - \frac{v^2}{c^2}\right)^{-3/2}.
$$

Since
$$
\gamma_{\text{velocity}} 
= \left(1 - \frac{v^2}{c^2}\right)^{-1/2},
\quad\Rightarrow\quad
\gamma_{\text{velocity}}^3
= \left(1 - \frac{v^2}{c^2}\right)^{-3/2},
$$

we have

$$
\frac{d\gamma_{\text{velocity}}}{dv}
= \frac{v}{c^2}\,\gamma_{\text{velocity}}^3,
\qquad
d\gamma_{\text{velocity}}
= \frac{v}{c^2}\,\gamma_{\text{velocity}}^3\,dv.
$$

$$
\textbf{4. Plug this into }dK
$$

Start from

$$
dK 
= m\gamma_{\text{life}}\Big(\gamma_{\text{velocity}} v\,dv 
  + v^2 d\gamma_{\text{velocity}}\Big).
$$

Insert
$$
d\gamma_{\text{velocity}}
= \frac{v}{c^2}\,\gamma_{\text{velocity}}^3\,dv:
$$

$$
\begin{aligned}
dK 
&= m\gamma_{\text{life}}\left(
    \gamma_{\text{velocity}} v\,dv
    + v^2 \frac{v}{c^2}\gamma_{\text{velocity}}^3\,dv
  \right) \\
&= m\gamma_{\text{life}}
   \left(
      \gamma_{\text{velocity}} v
      + \frac{v^3}{c^2}\gamma_{\text{velocity}}^3
   \right) dv.
\end{aligned}
$$

Factor out \(v\gamma_{\text{velocity}}\):

$$
dK
= m\gamma_{\text{life}}\,v\gamma_{\text{velocity}}
   \left(
      1 + \frac{v^2}{c^2}\gamma_{\text{velocity}}^2
   \right) dv.
$$

But

$$
\gamma_{\text{velocity}}^2
= \frac{1}{1 - \dfrac{v^2}{c^2}},
$$

so

$$
1 + \frac{v^2}{c^2}\gamma_{\text{velocity}}^2
= 1 + \frac{v^2}{c^2}\frac{1}{1 - \dfrac{v^2}{c^2}}
= \frac{1 - \dfrac{v^2}{c^2} + \dfrac{v^2}{c^2}}{1 - \dfrac{v^2}{c^2}}
= \gamma_{\text{velocity}}^2.
$$

Therefore

$$
dK
= m\gamma_{\text{life}}\,v\gamma_{\text{velocity}}^3\,dv.
$$

Using
$$
d\gamma_{\text{velocity}}
= \frac{v}{c^2}\gamma_{\text{velocity}}^3\,dv
\quad\Rightarrow\quad
v\gamma_{\text{velocity}}^3\,dv = c^2\,d\gamma_{\text{velocity}},
$$

we get

$$
dK = m\gamma_{\text{life}}\,c^2\,d\gamma_{\text{velocity}}.
$$

$$
\boxed{
dK = m\gamma_{\text{life}}\,c^2\,d\gamma_{\text{velocity}}
}
$$

$$
\textbf{5. Integrate from rest to some velocity}
$$

At rest:
$$
v = 0 \quad\Rightarrow\quad
\gamma_{\text{velocity}} = 1,
\qquad
K = 0.
$$

Integrate:

$$
\int_0^{K} dK
= m\gamma_{\text{life}}\,c^2
\int_{1}^{\gamma_{\text{velocity}}} d\gamma_{\text{velocity}}.
$$

So

$$
K = m\gamma_{\text{life}}\,c^2
\left[\gamma_{\text{velocity}}\right]_{1}^{\gamma_{\text{velocity}}}
= m\gamma_{\text{life}}\,c^2
\big(\gamma_{\text{velocity}} - 1\big).
$$

$$
\boxed{
K = m\gamma_{\text{life}}\,c^2
\big(\gamma_{\text{velocity}} - 1\big),
\qquad
\gamma_{\text{velocity}}
= \frac{1}{\sqrt{1 - \dfrac{v^2}{c^2}}}.
}
$$

$$
\text{Total kinetic energy} 
= \gamma_{\text{life}} \times
\text{(standard SR kinetic energy)}.
$$


$$
\boxed{
K = \gamma_{\text{life}}\,
\big(\gamma_{\text{velocity}} - 1\big)mc^2,
\qquad
\gamma_{\text{velocity}}
= \frac{1}{\sqrt{1 - \dfrac{v^2}{c^2}}}.
}
$$

$$
\textbf{Non-constant }\gamma_{\text{life}}
$$

$$
p = \Gamma\, m v,
\qquad
\Gamma = \gamma_{\text{life}}\,\gamma_{\text{velocity}}.
$$

Allow \(\gamma_{\text{life}}\) and \(\gamma_{\text{velocity}}\) \emph{both} to vary (with \(v\), or along the path):

$$
d\Gamma
= \gamma_{\text{velocity}}\,d\gamma_{\text{life}}
+ \gamma_{\text{life}}\,d\gamma_{\text{velocity}}.
$$

Workâ€“energy relation as before:

$$
dK = v\,dp,
\qquad
dp = m(\Gamma\,dv + v\,d\Gamma).
$$

So

$$
dK 
= v\,dp
= m\big(\Gamma v\,dv + v^2 d\Gamma\big).
$$

Insert \(\Gamma = \gamma_{\text{life}}\gamma_{\text{velocity}}\) and \(d\Gamma\) from above:

$$
\begin{aligned}
dK
&= m\Big(
   \gamma_{\text{life}}\gamma_{\text{velocity}} v\,dv
   + v^2\big(
       \gamma_{\text{velocity}}\,d\gamma_{\text{life}}
       + \gamma_{\text{life}}\,d\gamma_{\text{velocity}}
     \big)
   \Big) \\
&= m\Big(
    \gamma_{\text{life}}\gamma_{\text{velocity}} v\,dv
  + v^2\gamma_{\text{life}}\,d\gamma_{\text{velocity}}
  + v^2\gamma_{\text{velocity}}\,d\gamma_{\text{life}}
  \Big).
\end{aligned}
$$

Group the terms:

$$
dK
= m\gamma_{\text{life}}\Big(
      \gamma_{\text{velocity}} v\,dv
    + v^2 d\gamma_{\text{velocity}}
  \Big)
  \;+\;
  m v^2\gamma_{\text{velocity}}\,d\gamma_{\text{life}}.
$$

The bracket is the same combination as in the constant-\(\gamma_{\text{life}}\) case:

$$
\gamma_{\text{velocity}} v\,dv + v^2 d\gamma_{\text{velocity}}
= c^2\,d\gamma_{\text{velocity}},
$$

so we still get

$$
dK
= m\gamma_{\text{life}} c^2\,d\gamma_{\text{velocity}}
  \;+\;
  m v^2\gamma_{\text{velocity}}\,d\gamma_{\text{life}}.
$$

So the general differential is

$$
\boxed{
dK
= m\gamma_{\text{life}} c^2\,d\gamma_{\text{velocity}}
  \;+\;
  m v^2\gamma_{\text{velocity}}\,d\gamma_{\text{life}}.
}
$$

Using the identity
$$
\frac{v^2}{c^2}
= 1 - \frac{1}{\gamma_{\text{velocity}}^{2}}
\quad\Rightarrow\quad
v^2\gamma_{\text{velocity}}
= c^2\left(\gamma_{\text{velocity}} - \frac{1}{\gamma_{\text{velocity}}}\right),
$$

we can write the second term purely in terms of gammas:

$$
m v^2\gamma_{\text{velocity}}\,d\gamma_{\text{life}}
= m c^2
  \left(\gamma_{\text{velocity} }
        - \frac{1}{\gamma_{\text{velocity}}}
  \right)d\gamma_{\text{life}}.
$$

So an equivalent compact form is

$$
\boxed{
dK
= m c^2\Big[
      \gamma_{\text{life}}\,d\gamma_{\text{velocity}}
    + \Big(\gamma_{\text{velocity}}
           - \frac{1}{\gamma_{\text{velocity}}}
      \Big)d\gamma_{\text{life}}
  \Big].
}
$$

$$
\text{If }d\gamma_{\text{life}} = 0
\text{ we recover }
dK = m\gamma_{\text{life}} c^2 d\gamma_{\text{velocity}}.
$$
