# ⭐ **CLAUSIUS INEQUALITY — FULL EXPLANATION (LATEX FORMAT)**

# -----------------------------------------

## **1. Clausius Inequality Statement**

$$
\oint \frac{dQ}{T} \le 0
$$

* For **reversible cycles**:
  $$
  \oint \frac{dQ}{T} = 0
  $$

* For **irreversible cycles**:
  $$
  \oint \frac{dQ}{T} < 0
  $$

---

## **2. The Setup of the Cycle**

Clausius considers a **combined system** made of:

1. The **system**, which receives differential heat (dQ) at boundary temperature (T).
2. A **reversible cyclic device** connected to a reservoir at temperature (T_R).
   It exchanges differential heat (dQ_R) with the reservoir.

For the reversible device:

$$
\frac{dQ_R}{T_R} = \frac{dQ}{T}
$$

The combined system performs total differential work:

$$
dW_C = dW_{\text{rev}} + dW_{\text{sys}}
$$

Energy balance gives:

$$
dW_C = dQ_R - dE_C
$$

Over a full cycle:

$$
\oint dE_C = 0
$$

Thus:

$$
W_C = T_R \oint \frac{dQ}{T}
$$

---

## **3. Kelvin–Planck Argument**

The combined system exchanges heat *only* with a single reservoir at (T_R).
Kelvin–Planck states:

> No system can deliver net work while exchanging heat with a single reservoir.

Therefore:

$$
W_C \le 0
$$

Because (T_R > 0), we must have:

$$
\oint \frac{dQ}{T} \le 0
$$

---

## **4. Reversible Case**

If all steps are reversible:

* The cycle can be run backward.
* Kelvin–Planck forbids (W_C > 0) in forward direction,
  and also forbids (W_C < 0) in backward direction.

Thus:

$$
W_C = 0
$$

Which implies:

$$
\oint \frac{dQ}{T} = 0
$$

This defines entropy:

$$
dS = \left( \frac{dQ}{T} \right)_{\text{rev}}
$$

# -----------------------------------------

# ⭐ **NUMERICAL EXAMPLES (LATEX FORMAT)**

# -----------------------------------------

# ===============================

# ✅ **EXAMPLE 1 — REVERSIBLE CYCLE**

# ===============================

Heat engine operating between:

* Hot reservoir:
  $$
  T_H = 600 ,\text{K}
  $$
* Cold reservoir:
  $$
  T_C = 300 ,\text{K}
  $$

Let absorbed heat:

$$
Q_H = +600 \text{ kJ}
$$

### Efficiency of a reversible engine:

$$
\eta_{\text{rev}}
= 1 - \frac{T_C}{T_H}
= 1 - \frac{300}{600}
= 0.5
$$

### Work output:

$$
W = \eta_{\text{rev}} Q_H
= 0.5 \times 600
= 300 \text{ kJ}
$$

### Heat rejected:

$$
Q_C = Q_H - W = 600 - 300 = 300 \text{ kJ}
$$

(Sign convention:
Heat out of the system is negative.)

$$
Q_C = -300 \text{ kJ}
$$

### Clausius integral:

$$
\oint \frac{dQ}{T}
= \frac{Q_H}{T_H} + \frac{Q_C}{T_C}
$$

Substitute numbers:

$$
\oint \frac{dQ}{T}
= \frac{600}{600}

* \frac{-300}{300}
  = 1 - 1 = 0
  $$

### **Result:**

$$
\boxed{\oint \frac{dQ}{T} = 0 \quad \text{(Reversible Cycle)}}
$$

---

# ===============================

# ✅ **EXAMPLE 2 — IRREVERSIBLE CYCLE**

# ===============================

Same reservoirs:

* $T_H = 600 ,\text{K}$
* $T_C = 300 ,\text{K}$

Still absorbs:

$$
Q_H = +600 \text{ kJ}
$$

But the real engine efficiency is worse:

$$
\eta_{\text{irr}} = 0.30
$$

### Work output:

$$
W = 0.30 \times 600 = 180 \text{ kJ}
$$

### Heat rejected:

$$
Q_C = Q_H - W = 600 - 180 = 420 \text{ kJ}
$$

Again applying sign convention:

$$
Q_C = -420 \text{ kJ}
$$

### Clausius integral:

$$
\oint \frac{dQ}{T}
= \frac{600}{600}

* \frac{-420}{300}
  $$

Compute:

$$
\frac{600}{600}=1
$$

$$
\frac{-420}{300}=-1.4
$$

Total:

$$
\oint \frac{dQ}{T}
= 1 - 1.4
= -0.4
$$

### **Result:**

$$
\boxed{\oint \frac{dQ}{T} = -0.4 \quad \text{(Irreversible Cycle)}}
$$

### Entropy generated:

$$
S_{\text{gen}}
= - \oint \frac{dQ}{T}
= 0.4 , \text{kJ/K}
$$

---

# -----------------------------------------

# ⭐ **FINAL SUMMARY (LATEX)**

# -----------------------------------------

### **Reversible cycle:**

$$
\oint \frac{dQ}{T} = 0
$$

### **Irreversible cycle:**

$$
\oint \frac{dQ}{T} < 0
$$

Entropy generation is:

$$
S_{\text{gen}} = -\oint \frac{dQ}{T}
$$
