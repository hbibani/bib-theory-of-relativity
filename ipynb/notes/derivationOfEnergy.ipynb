  The total energy of a system can be defined as comprising of the internal energy (microscopic) and the kinetic energy (macroscopic)

  $$E_{total} = U_{internal} + Ki_{energy} + Mass_{rest}$$

  | Category        | Description                                                                            | Components                                |
  |-----------------|----------------------------------------------------------------------------------------|--------------------------------------------|

  | **Microscopic** | Processes related to molecular, energetic, and life-like configurational changes       | $p_{\text{energy}},\ p_{\text{life}}$     |

Starting from the total work expression,

$$
dW = dK + dU + dL
$$

and requiring consistency with the relativistic total energy,

$$
dE_{\text{total}} = d(\gamma m_{\text{eff}} c^2),
$$

we equate

$$
dK + dU + dL = d(\gamma m_{\text{eff}} c^2).
$$

Because the kinetic term already matches the relativistic expression,
the remaining terms must satisfy

$$
dU + dL = c^2 \, dm_{\text{eff}}.
$$

Therefore,

$$
dm_{\text{eff}} = \frac{dU}{c^2} + \frac{dL}{c^2}.
$$

Integrating,

$$
m_{\text{eff}} 
= m_0 
+ \frac{U}{c^2}
+ \frac{L}{c^2}.
$$

Thus the total relativistic energy becomes

$$
E_{\text{total}}
=
\gamma \left(
m_0 c^2 + U + L
\right).
$$

Effectively Formula turns out:

$$
p_{\text{matter}} = \sqrt{1-\frac{2G(M_{0} + \frac{U}{c^2} + \frac{L}{c^2})}{rc^2}}
$$

\textbf{Step 1 — Total work decomposition}

Starting point:

$$
dW = dK + dU + dL.
$$

The terms represent macroscopic kinetic work, microscopic internal work,
and life/configurational work.

\textbf{Step 2 — Kinetic work from relativistic momentum}

Relativistic momentum is

$$
p = \gamma m_{\text{eff}} v,
$$

so the kinetic work differential is

$$
dK = v \, dp.
$$

\textbf{Step 3 — Differentiate the momentum}

$$
dp = \frac{d}{dv}(\gamma m_{\text{eff}} v) \, dv
$$

Compute the differential:

$$
dp = \gamma m_{\text{eff}} \, dv
      + v \, m_{\text{eff}} \, d\gamma
      + v \gamma \, dm_{\text{eff}}.
$$

Thus

$$
dK = v\,dp
   = \gamma m_{\text{eff}} v\,dv
   + m_{\text{eff}} v^2 d\gamma
   + \gamma v^2 dm_{\text{eff}}.
$$

\textbf{Step 4 — Total energy must match the relativistic form}

Special relativity requires

$$
dE_{\text{total}}
= d(\gamma m_{\text{eff}} c^2).
$$

Differentiate:

$$
d(\gamma m_{\text{eff}} c^2)
=
c^2\left(
m_{\text{eff}}\,d\gamma
+
\gamma\,dm_{\text{eff}}
\right).
$$

\textbf{Step 5 — Equate the mechanical and relativistic expressions}

Since total work equals total energy input:

$$
dK + dU + dL
=
d(\gamma m_{\text{eff}} c^2).
$$

Substitute $dK$:

\begin{aligned}
\gamma m_{\text{eff}} v\,dv
&+ m_{\text{eff}} v^2 d\gamma
+ \gamma v^2 dm_{\text{eff}}
+ dU + dL
\\[6pt]
&=
c^2\left(
m_{\text{eff}} d\gamma
+
\gamma dm_{\text{eff}}
\right).
\end{aligned}

\textbf{Step 6 — Match the coefficients for all possible velocities}

The terms involving $dv$ and $d\gamma$
match the standard relativistic kinetic-energy derivation,
so the remaining unmatched part is:

$$
dU + dL = c^2 \, dm_{\text{eff}}.
$$

\textbf{Step 7 — Solve for effective mass}

$$
dm_{\text{eff}} = \frac{dU}{c^2} + \frac{dL}{c^2}.
$$

Integrate:

$$
m_{\text{eff}}
=
m_0
+
\frac{U}{c^2}
+
\frac{L}{c^2}.
$$

\textbf{Step 8 — Substitute this into total relativistic energy}

$$
E_{\text{total}}
=
\gamma m_{\text{eff}} c^2
=
\gamma ( m_0 c^2 + U + L ).
$$

### Microscopic Kinetic-Energy Derivation of Internal-Energy Mass

### Step 1 — Internal energy as microscopic kinetic energy

For $N$ microscopic particles, each with rest mass $m_0$ and velocity $v_i$,
the relativistic kinetic energy of each particle is

$$
K_i = (\gamma_i - 1) m_0 c^2.
$$

Thus the total internal energy is

$$
U = \sum_{i=1}^{N} (\gamma_i - 1) m_0 c^2.
$$


### Step 2 — Microscopic momentum and differential work

The momentum of particle $i$ is

$$
p_i = \gamma_i m_0 v_i.
$$

Work done on microscopic particles contributes to internal energy:

$$
dU = \sum_i v_i \, dp_i.
$$


### Step 3 — Differentiate microscopic momentum

$$
dp_i = \frac{d}{dv_i}(\gamma_i m_0 v_i)\, dv_i.
$$

Apply the product rule:

$$
dp_i
=
\gamma_i m_0 \, dv_i
+
v_i m_0 \, d\gamma_i.
$$

Thus

$$
dU
=
\sum_i
v_i \, dp_i
=
\sum_i \left(
\gamma_i m_0 v_i \, dv_i
+
m_0 v_i^2 \, d\gamma_i
\right).
$$


###  Step 4 — Use the key relativistic identity

For each particle,

$$
\gamma_i v_i\, dv_i + v_i^2 d\gamma_i = c^2 \, d\gamma_i.
$$

Therefore,

$$
dU = \sum_i m_0 c^2 \, d\gamma_i.
$$

Integrate:

$$
U = \sum_i m_0 c^2 (\gamma_i - 1).
$$


### Step 5 — Compare internal energy to rest-energy form

Total rest-frame energy:

$$
E_{\text{rest}}
=
m_0 c^2 + U.
$$

Special relativity requires that rest-frame energy always satisfies

$$
E_{\text{rest}} = m_{\text{eff}} c^2.
$$

Equate the expressions:

$$
m_{\text{eff}} c^2
=
m_0 c^2 + U.
$$

Solve for $m_{\text{eff}}$:

$$
m_{\text{eff}}
=
m_0 + \frac{U}{c^2}.
$$

### Conclusion

Microscopic kinetic energy increases the effective mass by

$$
\boxed{
m_{\text{eff}} = m_0 + \frac{U}{c^2}.
}
$$
