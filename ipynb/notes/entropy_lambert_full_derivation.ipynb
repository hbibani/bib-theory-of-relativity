
# Entropy–Lambert W Dilation: Full Explicit Derivation  

---

# 1. Entropy Definition

For a system with \(N\) boxes and probability distribution  
\(\{p_1, p_2, \dots, p_N\}\), the Shannon entropy is:

$$
S \;=\; \sum_{i=1}^{N} p_i \ln p_i
$$

---

# 2. Isolating the Contribution of a Single Box

Suppose we are solving for box \(j\).  
Define:

$$
S_{\setminus j} \;=\; \sum_{i \ne j} p_i \ln p_i
$$

Then:

$$
S \;=\; p_j \ln p_j \;+\; S_{\setminus j}
$$

Rearrange to isolate the unknown term:

$$
p_j \ln p_j 
\;=\;
S - S_{\setminus j}
$$

Define the helper quantity:

$$
K \;\equiv\; S - S_{\setminus j}
$$

So we have:

$$
p_j \ln p_j = K
$$

---

# 3. Solving the Equation \(p \ln p = K\)

Let:

$$
p_j = e^x
$$

Then:

$$
\ln p_j = x
$$

Substitute into the original equation:

$$
p_j \ln p_j = e^x \cdot x = K
$$

Thus the equation reduces to:

$$
x e^x = K
$$

The Lambert W function is defined as the inverse of:

$$
W(z) e^{W(z)} = z
$$

Therefore:

$$
x = W(K)
$$

Recovering \(p_j\):

$$
p_j = e^{W(K)}
$$

Using the identity:

$$
e^{W(K)} = \frac{K}{W(K)}
$$

We obtain the exact inversion:

$$
p_j = \frac{K}{W(K)}
$$

## 2. Simple example: one box with (p = 0.5)

Let’s ignore the rest of the boxes and just test the pure inversion:

Take
$$
p = 0.5.
$$

Compute

$$
K = p \ln p = 0.5 \ln(0.5).
$$

We know
$$
\ln(0.5) = -\ln 2 \approx -0.693147,
$$

so

$$
K = 0.5 \ln(0.5)
= 0.5 \cdot (-0.693147)
\approx -0.3465736.
$$

So:

* True probability: (p_{\text{true}} = 0.5).
* Computed (K \approx -0.3465736).

Now we apply the Lambert inversion:

> Solve (p \ln p = K) for (p).

### 2.1 Why (p = K / W(K))?

Start from
$$
p \ln p = K.
$$

Let
$$
x = \ln p
\quad\Rightarrow\quad
p = e^x.
$$

Then
$$
p \ln p = e^x \cdot x = x e^x = K.
$$

By **definition of the Lambert W function**, if (x e^x = K), then
$$
x = W(K).
$$

So
$$
\ln p = W(K)
\quad\Rightarrow\quad
p = e^{W(K)}.
$$

But also from (K = p \ln p):

$$
K = p \ln p
= p, W(K)
\quad\Rightarrow\quad
p = \frac{K}{W(K)}.
$$

So we have **two equivalent forms**:
$$
p = e^{W(K)}
\quad\text{and}\quad
p = \frac{K}{W(K)}.
$$

---

### 2.2 Plug in the numbers

We had (K \approx -0.3465736).

For this (K),

* The Lambert W satisfies
  $$W(K), e^{W(K)} = K.$$
* In fact, for this specific (K), the solution is
  $$W(K) = \ln(0.5) = -0.693147\ldots$$
  (you can verify: (\ln(0.5) \cdot 0.5 = -0.3465736 = K)).

Now compute

$$
p = \frac{K}{W(K)}
= \frac{-0.3465736}{-0.693147}
= 0.5.
$$

So the inversion works **exactly**:

* We started with (p = 0.5),
* computed (K = p \ln p),
* then recovered (p) from
  $$p = \frac{K}{W(K)}$$
  and got back (0.5).

## 2. Boltzmann multiplicity for this macrostate

Let there be (N) molecules, with occupancies

[
n_1 = \frac{N}{2}, \quad
n_2 = n_3 = n_4 = n_5 = \frac{N}{8}.
]

Assume (N) is a multiple of 8.

The Boltzmann multiplicity for 5 boxes is

[
W = \frac{N!}{n_1!,n_2!,n_3!,n_4!,n_5!}
=======================================

\frac{N!}{\left(\frac{N}{2}\right)!\left(\frac{N}{8}\right)!^4}.
]

So

[
\ln W
= \ln N!

* \ln\left(\frac{N}{2}!\right)
* 4 \ln\left(\frac{N}{8}!\right).
  ]

Now use Stirling’s approximation for large (N):

[
\ln m! \approx m\ln m - m.
]

Then

[
\begin{aligned}
\ln W
&\approx \left[ N\ln N - N \right]\
&\quad
-\left[\frac{N}{2}\ln\left(\frac{N}{2}\right) - \frac{N}{2}\right]
-4\left[\frac{N}{8}\ln\left(\frac{N}{8}\right) - \frac{N}{8}\right].
\end{aligned}
]

Work it out:

* First, the “(-m)” pieces:
  [
  -N + \frac{N}{2} + 4\cdot\frac{N}{8} = -N + \frac{N}{2} + \frac{N}{2} = 0.
  ]
  So those cancel completely.

* We are left with the log terms:
  [
  \ln W \approx
  N \ln N

  * \frac{N}{2}\ln\left(\frac{N}{2}\right)
  * 4\cdot\frac{N}{8}\ln\left(\frac{N}{8}\right).
    ]

But (4 \cdot \frac{N}{8} = \frac{N}{2}), so

[
\ln W
\approx
N\ln N

* \frac{N}{2}\ln\left(\frac{N}{2}\right)
* \frac{N}{2}\ln\left(\frac{N}{8}\right).
  ]

Factor out (\frac{N}{2}):

[
\ln W
\approx
\frac{N}{2}
\left[
2\ln N

* \ln\left(\frac{N}{2}\right)
* \ln\left(\frac{N}{8}\right)
  \right].
  ]

Combine the logs:

[
\ln\left(\frac{N}{2}\right) + \ln\left(\frac{N}{8}\right)
= \ln\left(\frac{N^2}{16}\right)
= 2\ln N - \ln 16.
]

So:

[
\begin{aligned}
\ln W
&\approx \frac{N}{2}\left[2\ln N - (2\ln N - \ln 16)\right][4pt]
&= \frac{N}{2} \cdot \ln 16
= N \ln 4.
\end{aligned}
]

So **Boltzmann** gives (to leading order):

[
\ln W \approx N \ln 4.
]

Thus physical entropy:

[
S_{\text{Boltz}}
= k_B \ln W
\approx k_B N \ln 4.
]

So **per particle**:

[
\frac{S_{\text{Boltz}}}{k_B N}
\approx \ln 4.
]

This is the usual “positive entropy” sign.

## 3. Match this to your (S = \sum p_i \ln p_i)

From the probability side:

[
\begin{aligned}
\sum_{i=1}^5 p_i \ln p_i
&= \frac{1}{2}\ln\left(\frac{1}{2}\right)

* 4 \cdot \frac{1}{8}\ln\left(\frac{1}{8}\right)[4pt]
  &= \frac{1}{2}\ln\left(\frac{1}{2}\right)
* \frac{1}{2}\ln\left(\frac{1}{8}\right)[4pt]
  &= \frac{1}{2}
  \left[
  \ln\left(\frac{1}{2}\right)
* \ln\left(\frac{1}{8}\right)
  \right][4pt]
  &= \frac{1}{2}\ln\left(\frac{1}{16}\right)
  = -\frac{1}{2}\ln 16
  = -\ln 4.
  \end{aligned}
  ]

So:

[
S_{\text{math}} \equiv \sum p_i \ln p_i = -\ln 4.
]

Compare with Boltzmann’s result:

[
\frac{S_{\text{Boltz}}}{k_B N} \approx \ln 4.
]

So they are related by

[
S_{\text{math}} = -\frac{S_{\text{Boltz}}}{k_B N}.
]

That is: your “(S)” is just the **negative per-particle Boltzmann entropy** (which is exactly how the Shannon form emerges from Boltzmann).

So **yes**, using the general ( \ln W) formula and Stirling, we recover the same (\sum p_i \ln p_i) (up to the overall minus sign and factor (k_B N)), and hence the *same* (K).

## 4. Now compute (K) and check the Lambert inversion

You defined

[
K = S_{\text{math}} - \sum_{i=2}^5 p_i \ln p_i.
]

We already have:

* All-box sum:
  [
  S_{\text{math}} = -\ln 4.
  ]
* Sum over boxes 2–5:
  [
  \sum_{i=2}^5 p_i \ln p_i
  = 4\cdot\frac{1}{8}\ln\left(\frac{1}{8}\right)
  = \frac{1}{2}\ln\left(\frac{1}{8}\right)
  = -\frac{1}{2}\ln 8.
  ]

So

[
\begin{aligned}
K
&= -\ln 4 -\left(-\frac{1}{2}\ln 8\right)[4pt]
&= -\ln 4 + \frac{1}{2}\ln 8[4pt]
&= \frac{1}{2}\left(\ln 8 - 2\ln 4\right)[4pt]
&= \frac{1}{2}\left(\ln 8 - \ln 16\right)[4pt]
&= \frac{1}{2}\ln\left(\frac{8}{16}\right)
= \frac{1}{2}\ln\left(\frac{1}{2}\right).
\end{aligned}
]

But
[
p_1 \ln p_1
= \frac{1}{2}\ln\left(\frac{1}{2}\right),
]
so

[
K = p_1 \ln p_1,
]
exactly as intended.

Numerically:

* (\ln(1/2) \approx -0.693147)
* (K = \tfrac{1}{2}\ln(1/2) \approx -0.3465736).

From the Lambert inversion:

[
p_1 = \frac{K}{W(K)}.
]

Here (K = p_1\ln p_1), so (W(K) = \ln p_1), and

[
p_1 = \frac{K}{W(K)}
= \frac{p_1 \ln p_1}{\ln p_1}
= p_1,
]
which in this example is (p_1 = 0.5).

So:

* **Boltzmann** (S = k_B \ln W) with Stirling gives
  (\displaystyle \frac{S_{\text{Boltz}}}{k_B N} \approx \ln 4).
* Your **math entropy** (S_{\text{math}} = \sum p_i \ln p_i) is
  (\displaystyle S_{\text{math}} = -\ln 4).
* Using that (S_{\text{math}}) in your
  (\displaystyle K = S_{\text{math}} - \sum_{i=2} p_i \ln p_i)
  gives (K = p_1 \ln p_1).
* The Lambert formula
  (\displaystyle p_1 = \frac{K}{W(K)})
  then recovers the correct (p_1).

So the number you get from the **general Boltzmann (\ln W)** (after dividing by (k_B N) and flipping the sign) is *exactly the same* object feeding into your (K) and Lambert inversion. It “works” and is consistent with the combinatorial definition.
