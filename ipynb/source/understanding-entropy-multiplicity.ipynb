In computational relativity, probability dilation was expressed in the following form:

$$
p_{\mathrm{obs}}
=
\gamma p_{\mathrm{exp}}
=

$$

This relation can be rearranged into the following energy form:

$$
\Delta E

In this framework, the expression can be interpreted as probability dilation reducing the effective disorder-producing energy within the system. The structure of this term is closely related to Boltzmann’s entropy relation:

$$

$$

This suggests that probability dilation interacts with the Boltzmann distribution in a way that may lower the entropy of the system by reducing the number of accessible disorder-producing configurations.

# Entropy as a portion of the system's total internal energy


**Boltzmann Entropy System**
The total energy of a microscopic system within a thermodynamic environment is represented by the internal energy term:

$$
U
$$

This internal energy can include different forms of energy, depending on the physical processes occurring in the system and the type of work being performed. In the following table, entropy can be interpreted as a form of disorder-producing work. Unlike directed mechanical work, such as a heat engine using energy to move a fan, entropy describes work that redistributes energy into many microscopic configurations. This type of work does not produce organized motion; instead, it increases the disorder of the system.

This can be represented by the entropy-energy term:

$$
T\Delta S
$$

where $T\Delta S$ represents the part of the system’s energy associated with entropy production. In this interpretation, directed work produces organized motion, while entropy-producing work disperses energy into disorder.


| Summary                      | Mathematical Form                                                      | Meaning                                                                                                                                                                                                                                                            |
| ---------------------------- | ---------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **Boltzmann entropy system** | $$U = E_{\mathrm{other}} + T\Delta S, \qquad S = k_B \ln \Omega$$ | The internal energy $U$ contains energy that can perform directed work and energy that becomes dispersed into disorder-producing microscopic configurations. Entropy describes the dispersed, configuration-based part of the system, not the total energy itself. |
| **Work done on a fan**       | $$U = E_{\mathrm{other}} + E_{\mathrm{work}} + T\Delta S$$                                     | When energy from the system is used to move a fan, part of the internal energy is converted into directed mechanical work. This work is directive because the energy is used to produce organized motion rather than being dispersed into disorder. The work done by the engine is only a portion of the total energy of the system.              |

# Probability Dilation and the effect anti-entropy has on the total entropy of the system

With the addition of probability dilation, we can introduce an anti-entropy term that represents the conversion of disordered entropic energy into ordered energy. This reduction in effective entropy is caused by probability dilation, where the factor (\gamma) modifies the energy relation:

$$
\Delta E
=
-k_B T \ln \gamma
$$

This can be interpreted as an entropy-lowering energy contribution. We define the anti-entropy term as:

$$
L
=
k_B \ln \gamma
$$

where (L) represents the entropy-reducing component associated with probability dilation. We also include the ordinary entropy contribution of the system:

$$
S
=
k_B \ln \Omega
$$

Therefore, when probability dilation is present, the remaining disordered entropic contribution is:

$$
E_{\mathrm{t}}
=
T\Delta S
-
T\Delta L
$$


Therefore, when the energy in the system is non-constant, the energy of the system is reduced and the total internal energy can be written as:

$$
U_{nc}
=
E_{\mathrm{other}}
+
\left(T\Delta S-T\Delta L\right)
$$

When the energy in the system is constant, the total energy can be written as:

$$
U_{c}
=
E_{\mathrm{other}}
+
\left(T\Delta S-T\Delta L\right) + T\Delta L
$$

It is important to understand that, in the expression above, cancelling the anti-entropy term changes the interpretation of the equation. Without cancellation, the equation implies that the system contains two distinct energy contributions: one associated with organized energy and one associated with disorganized energy. The entropy of the system has been reduced and converted by the computational field.

# Redefining Multiplicity Through Internal Energy to Describe Entropy as a Component of Total Energy


The total energy of the system can be retrieved by interpreting multiplicity in this form


$$
\Omega(U,V,N)
=

C V^N U^{\frac{3N}{2}}
$$

This shows that entropy is related to the total internal energy of the system, but it should not be treated as the total energy itself. Rather, entropy describes that a portion of the total energy is used to produce disorder. This description shows how entropy is related to the distribution of internal energy within the system, and it can also be connected to the energy relation:

$$
E = \frac{3}{2}N k_B T
$$

In this sense, entropy is a part of the system’s energy description (not the entire energy of the system), and can be defined as the work done that produces disorder. Understanding probability dilation in this framework suggests that a computational field may reduce this disorder, thereby lowering the entropy of the system.


Starting from:

$$
S = k_B \ln \Omega
$$

we substitute the multiplicity definition:

$$
\Omega(U,V,N)
=

C V^N U^{\frac{3N}{2}}
$$

This gives:

$$
S
=

k_B
\ln
\left(
C V^N U^{\frac{3N}{2}}
\right)
$$

Which leads to the total energy of the system:

$$
U = \frac{3}{2}N k_B T
$$


The terms in this multiplicity expression can be described as follows:

| Quantity            |      Symbol     | Description                                                                                                                 | Relation to Multiplicity                                                                 |
| ------------------- | :-------------: | --------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------- |
| Multiplicity        | $\Omega(U,V,N)$ | Number of accessible microscopic configurations for a system with internal energy $U$, volume $V$, and particle number $N$. | Defined by $\Omega(U,V,N)=C V^N U^{\frac{3N}{2}}$.                                       |
| Internal energy     |       $U$       | Total microscopic energy available to the particles in the thermodynamic system.                                            | Increasing $U$ increases multiplicity through the factor $U^{\frac{3N}{2}}$.             |
| Volume              |       $V$       | Amount of space available for the particles to occupy.                                                                      | Increasing $V$ increases multiplicity through the factor $V^N$.                          |
| Number of particles |       $N$       | Total number of particles in the system.                                                                                    | Increasing $N$ increases the powers applied to both $V$ and $U$.                         |
| Constant factor     |       $C$       | A proportionality constant that contains fixed system-dependent factors.                                                    | Sets the scale of $\Omega$ but does not change the main dependence on $U$, $V$, and $N$. |
| Entropy             |       $S$       | Statistical measure of the number of accessible microstates.                                                                | Related to multiplicity by $S=k_B\ln\Omega$.                                             |
| Boltzmann constant  |      $k_B$      | Converts the logarithm of multiplicity into physical entropy units.                                                         | Sets the entropy scale.                                                                  |

As $N$ increases, the powers of both $V$ and $U$ increase:

$$
\Omega(U,V,N)
=============

C V^N U^{\frac{3N}{2}}
$$

Therefore, a larger number of particles produces a much larger number of possible microscopic arrangements.

| Number of particles $N$ | Multiplicity form                               | Entropy behavior    |
| ----------------------: | ----------------------------------------------- | ------------------- |
|                     $1$ | $\Omega = C V U^{\frac{3}{2}}$                  | Low entropy         |
|                    $10$ | $\Omega = C V^{10} U^{15}$                      | Larger entropy      |
|               $10^{23}$ | $\Omega = C V^{10^{23}} U^{\frac{3}{2}10^{23}}$ | Macroscopic entropy |

This shows why multiplicity grows extremely quickly in thermodynamic systems. Even small increases in the number of particles can produce a massive increase in the number of possible microscopic configurations.
