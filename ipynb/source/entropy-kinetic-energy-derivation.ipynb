---

# Deriving Total Energy from Boltzmann Entropy

We begin with the Boltzmann entropy relation:

$$
S = k_B \ln \Omega
$$

where \(S\) is the entropy, \(k_B\) is the Boltzmann constant, and \(\Omega\) is the multiplicity of the system.

For a thermodynamic system, the multiplicity depends on the internal energy, volume, and number of particles:

$$
\Omega = \Omega(U,V,N)
$$

Therefore, the Boltzmann entropy relation can be written more completely as:

$$
S = k_B \ln \Omega(U,V,N)
$$

---

# Step 1: Choose the system

We consider an ideal monatomic gas with \(N\) particles.

For an ideal monatomic gas, the internal energy is the total translational kinetic energy of all particles in the system:

$$
U = K_{\text{total}}
$$

Therefore:

$$
U
=
\sum_{i=1}^{N}
\frac{1}{2}m v_i^2
$$

Our goal is to derive:

$$
U
=
\frac{3}{2}Nk_BT
$$

from the Boltzmann entropy relation.

---

# Step 2: Write multiplicity as a function of energy

For an ideal monatomic gas, the multiplicity scales approximately as:

$$
\Omega(U,V,N)
\propto
V^N U^{\frac{3N}{2}}
$$

This can be written as:

$$
\Omega(U,V,N)
=
C V^N U^{\frac{3N}{2}}
$$

where \(C\) is a constant that contains factors independent of \(U\), such as mass terms, geometric factors, Planck's constant, and particle-counting corrections.

The important part for this derivation is the energy dependence:

$$
\Omega
\propto
U^{\frac{3N}{2}}
$$

---

# Step 3: Substitute multiplicity into Boltzmann entropy

Starting from:

$$
S = k_B \ln \Omega
$$

substitute:

$$
\Omega(U,V,N)
=
C V^N U^{\frac{3N}{2}}
$$

This gives:

$$
S
=
k_B
\ln
\left(
C V^N U^{\frac{3N}{2}}
\right)
$$

---

# Step 4: Expand the logarithm

Using the logarithm rule:

$$
\ln(abc)
=
\ln a
+
\ln b
+
\ln c
$$

we get:

$$
S
=
k_B
\left[
\ln C
+
\ln(V^N)
+
\ln
\left(
U^{\frac{3N}{2}}
\right)
\right]
$$

Now apply:

$$
\ln(V^N)
=
N\ln V
$$

and:

$$
\ln
\left(
U^{\frac{3N}{2}}
\right)
=
\frac{3N}{2}\ln U
$$

Therefore:

$$
S
=
k_B
\left[
\ln C
+
N\ln V
+
\frac{3N}{2}\ln U
\right]
$$

---

# Step 5: Use the thermodynamic definition of temperature

Temperature is defined by how entropy changes as internal energy changes:

$$
\frac{1}{T}
=
\left(
\frac{\partial S}{\partial U}
\right)_{V,N}
$$

This means that we differentiate entropy with respect to \(U\), while holding \(V\) and \(N\) constant.

So:

$$
\frac{1}{T}
=
\frac{\partial}{\partial U}
\left[
k_B
\left(
\ln C
+
N\ln V
+
\frac{3N}{2}\ln U
\right)
\right]
$$

---

# Step 6: Differentiate term by term

The entropy expression is:

$$
S
=
k_B
\left[
\ln C
+
N\ln V
+
\frac{3N}{2}\ln U
\right]
$$

Since \(C\) is constant:

$$
\frac{\partial}{\partial U}
(\ln C)
=
0
$$

Since \(V\) and \(N\) are held constant:

$$
\frac{\partial}{\partial U}
(N\ln V)
=
0
$$

Only the final term depends on \(U\):

$$
\frac{\partial}{\partial U}
\left(
\frac{3N}{2}\ln U
\right)
=
\frac{3N}{2}
\frac{1}{U}
$$

Therefore:

$$
\left(
\frac{\partial S}{\partial U}
\right)_{V,N}
=
k_B
\frac{3N}{2}
\frac{1}{U}
$$

So:

$$
\frac{1}{T}
=
\frac{3Nk_B}{2U}
$$

---

# Step 7: Solve for the internal energy

Starting from:

$$
\frac{1}{T}
=
\frac{3Nk_B}{2U}
$$

Invert both sides:

$$
T
=
\frac{2U}{3Nk_B}
$$

Now solve for \(U\):

$$
U
=
\frac{3}{2}Nk_BT
$$

Therefore:

$$
\boxed{
U
=
\frac{3}{2}Nk_BT
}
$$

---

# Final Result

Starting from Boltzmann entropy:

$$
S
=
k_B\ln\Omega(U,V,N)
$$

and using the ideal-gas multiplicity scaling:

$$
\Omega(U,V,N)
\propto
V^N U^{\frac{3N}{2}}
$$

we obtain:

$$
\boxed{
U
=
\frac{3}{2}Nk_BT
}
$$

This shows that Boltzmann entropy gives the total energy of the system indirectly. The energy is not obtained by multiplying \(S\) by \(T\). Instead, it is obtained from how the multiplicity changes with internal energy:

$$
\boxed{
\frac{1}{T}
=
\left(
\frac{\partial S}{\partial U}
\right)_{V,N}
}
$$

For an ideal monatomic gas, this gives:

$$
\boxed{
U
=
K_{\text{total}}
=
\frac{3}{2}Nk_BT
}
$$
