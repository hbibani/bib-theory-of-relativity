An organism capable of producing a UFO would, by virtue of its evolutionary development, have to be angelic. But if such an organism does not intervene to protect the earth, then its behavior stands in contradiction to its supposed angelic character.

There are two kinds of fields:

1. low-energy fields
2. high-energy fields

There are also two kinds of organisms:

1. those in the absolute reference frame
2. those below, or within the boundary of, that frame

So the creation of the UFO must come from one of two sources:

1. a high-energy field, or
2. an organism that exists below the boundary of the absolute reference frame

But if the producer’s behavior is inconsistent with what would be expected of an angelic organism, then it is more reasonable to conclude that the UFO is the product of a high-energy field.
