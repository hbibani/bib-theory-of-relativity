%==========================================================
% Energy removal -> speed/momentum reduction (non-relativistic)
% with e = E_total/(n k_e)
%==========================================================

%-----------------------------
% Definitions
%-----------------------------

$$
K=\frac{1}{2}mv^2,
\qquad
p=mv.
$$

%-----------------------------
% Remove energy e from kinetic energy
%-----------------------------

$$
K' = K - e.
$$

$$
\frac{1}{2}mv'^2 = \frac{1}{2}mv^2 - e.
$$

$$
mv'^2 = mv^2 - 2e
\quad\Longrightarrow\quad
v'^2 = v^2 - \frac{2e}{m}.
$$

%-----------------------------
% Express v' in factored form
%-----------------------------

$$
v'^2
=
v^2\left(1-\frac{2e}{mv^2}\right)
\quad\Longrightarrow\quad
v' = v\sqrt{1-\frac{2e}{mv^2}}.
$$

%-----------------------------
% Momentum after energy removal
%-----------------------------

$$
p' = mv' = m\sqrt{\,v^2-\frac{2e}{m}\,}.
$$

$$
p'^2
=
m^2\left(v^2-\frac{2e}{m}\right)
=
m^2v^2-2me.
$$

$$
\boxed{\,p'^2 = m^2v^2 - 2me\,}.
$$

Using \(p^2=m^2v^2\):

$$
\boxed{\,p'^2 = p^2 - 2me\,}.
$$

%==========================================================
% Parameterization: e = E_total/(n k_e)
%==========================================================

$$
e=\frac{E_{\text{total}}}{n\,k_e}.
$$

Let

$$
E_{\text{total}}=\frac{1}{2}mv^2 \equiv K,
\qquad
n=1.
$$

Then

$$
e=\frac{K}{k_e}.
$$

So the dimensionless ratio becomes

$$
\frac{2e}{mv^2}
=
\frac{e}{\frac{1}{2}mv^2}
=
\frac{e}{K}
=
\frac{1}{k_e}.
$$

Substitute into the speed relation:

$$
v'
=
v\sqrt{1-\frac{2e}{mv^2}}
=
v\sqrt{1-\frac{1}{k_e}}.
$$

$$
\boxed{\,v' = v\sqrt{1-\frac{1}{k_e}}\,}.
$$

Reality condition:

$$
1-\frac{1}{k_e}\ge 0
\quad\Longleftrightarrow\quad
\boxed{\,k_e\ge 1\,}.
$$
