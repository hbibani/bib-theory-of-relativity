%==========================================================
% Energy removal -> speed/momentum reduction (non-relativistic)
% with e = (E_total/k_e) * E_total
% i.e. e = E_total^2 / k_e
%==========================================================

%-----------------------------
% Definitions
%-----------------------------

$$
K=\frac{1}{2}mv^2,
\qquad
p=mv.
$$

%-----------------------------
% Parameterization of removed energy
%-----------------------------

$$
e=\frac{E_{\text{total}}}{k_e}\,E_{\text{total}}
=\frac{E_{\text{total}}^{2}}{k_e}.
$$

Let

$$
E_{\text{total}}=\frac{1}{2}mv^2 \equiv K.
$$

Then

$$
e=\frac{K^2}{k_e}.
$$

%-----------------------------
% Remove energy e from kinetic energy
%-----------------------------

$$
K' = K - e.
$$

$$
\frac{1}{2}mv'^2
=
\frac{1}{2}mv^2
-
e.
$$

$$
mv'^2
=
mv^2
-
2e
\quad\Longrightarrow\quad
v'^2
=
v^2-\frac{2e}{m}.
$$

%-----------------------------
% Substitute e = K^2/k_e
%-----------------------------

$$
v'^2
=
v^2
-
\frac{2}{m}\,\frac{K^2}{k_e}.
$$

Using

$$
K=\frac{1}{2}mv^2
\quad\Longrightarrow\quad
K^2=\left(\frac{1}{2}mv^2\right)^2=\frac{1}{4}m^2v^4,
$$

we get

$$
v'^2
=
v^2
-
\frac{2}{m}\cdot\frac{1}{k_e}\cdot\frac{1}{4}m^2v^4
=
v^2-\frac{1}{2}\frac{m v^4}{k_e}.
$$

%-----------------------------
% Express v' in factored form
%-----------------------------

$$
v'^2
=
v^2\left(1-\frac{1}{2}\frac{m v^2}{k_e}\right)
\quad\Longrightarrow\quad
v'
=
v\sqrt{1-\frac{1}{2}\frac{m v^2}{k_e}}.
$$

$$
\boxed{\,v' = v\sqrt{1-\frac{m v^2}{2k_e}}\,}.
$$

%-----------------------------
% Momentum after energy removal
%-----------------------------

$$
p' = mv'
= mv\sqrt{1-\frac{m v^2}{2k_e}}.
$$

$$
p'^2
=
m^2v^2\left(1-\frac{m v^2}{2k_e}\right)
=
m^2v^2-\frac{m^3v^4}{2k_e}.
$$

Using \(p^2=m^2v^2\) and \(p^4=m^4v^4\Rightarrow v^4=p^4/m^4\), we can also write

$$
p'^2
=
p^2-\frac{m^3}{2k_e}\cdot\frac{p^4}{m^4}
=
p^2-\frac{p^4}{2mk_e}.
$$

$$
\boxed{\,p'^2 = p^2-\frac{p^4}{2mk_e}\,}.
$$

%-----------------------------
% Reality/positivity condition
%-----------------------------

For real \(v'\), we require

$$
1-\frac{m v^2}{2k_e}\ge 0
\quad\Longleftrightarrow\quad
\boxed{\,k_e\ge \frac{1}{2}m v^2\,}.
$$

# Derivation of Lorrentz Transformation using processing factor
---
$$
\frac{1}{2}m{v'_0}^{\,2}
=
\frac{1}{2}mv^2
-
\left(\frac{v^2}{c^2}\right)mv^2
$$

$$
\text{Cancel }m\ (m\neq 0):
\qquad
\frac{1}{2}{v'_0}^{\,2}
=
\frac{1}{2}v^2
-
\frac{v^4}{c^2}
$$

$$
\text{Multiply both sides by }2:
\qquad
{v'_0}^{\,2}
=
v^2
-
\frac{2v^4}{c^2}
=
v^2\left(1-\frac{2v^2}{c^2}\right)
$$

$$
\Rightarrow\quad
v'_0
=
v\sqrt{1-\frac{2v^2}{c^2}}
$$

$$
\textbf{Note: }\text{To derive }v' = v\sqrt{1-\frac{v^2}{c^2}}\text{ exactly, the equation must be}
$$

$$
\frac{1}{2}mv'^2
=
\frac{1}{2}mv^2
-
\frac{1}{2}\left(\frac{v^2}{c^2}\right)mv^2
$$

$$
\text{Cancel }m,\ \text{multiply by }2:
\qquad
v'^2
=
v^2
-
\frac{v^4}{c^2}
=
v^2\left(1-\frac{v^2}{c^2}\right)
$$

$$
\Rightarrow\quad
v'
=
v\sqrt{1-\frac{v^2}{c^2}}
$$

$$
\gamma
=
\frac{1}{\sqrt{1-\frac{v^2}{c^2}}}
\qquad\Rightarrow\qquad
v'=\frac{v}{\gamma}
$$

$$
\frac{mv'}{\sqrt{1-\frac{v^2}{c^2}}}
=
mv
\qquad\Longleftrightarrow\qquad
\gamma mv' = mv
$$

$$
\text{Verification: }
\qquad
\frac{mv'}{\sqrt{1-\frac{v^2}{c^2}}}
=
\frac{m\left(v\sqrt{1-\frac{v^2}{c^2}}\right)}{\sqrt{1-\frac{v^2}{c^2}}}
=
mv
$$
