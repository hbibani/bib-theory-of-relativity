## COMPARISON OF AGREEMENT PATTERNS



| Case | Chosen reference | If 100 s elapsed there, do all other observers/frames recover the same 100 s relative to that reference? | Do they agree with each other locally on the elapsed time? | Main reason |
| --- | --- | --- | --- | --- |
| Lorentz / SR with absolute inertial reference frame | Absolute inertial reference frame | Yes | No | All moving observers can transform back to the same inertial reference frame and recover the same elapsed time there, but their own proper times differ |
| Schwarzschild with infinity as reference | Static observer at infinity | Yes | No | All static observers at finite radii can reconstruct the same elapsed time at infinity, but their own local proper times differ by radius |
| Lorentz / SR with a moving frame chosen as reference (e.g. \(0.9c\)) | A moving observer at \(0.9c\) | No, not in the same universal way as for the absolute inertial frame | No | Other moving observers do not generally share one common recovered elapsed time in that moving frame in the same reference-frame sense; pairwise relative times differ |
| Schwarzschild with a finite radius chosen as reference | A finite radius \(r_A\), not infinity | No, not in the same universal way as for infinity | No | Other radii do not generally share one common local elapsed time relative to that finite-radius clock; pairwise elapsed times differ by radius ratio |

---

# FORMULAS USED

| Case | Reference chosen | Total formula |
| --- | --- | --- |
| 1. Lorentz / SR | Absolute inertial frame | $\Delta \tau(v)=100\sqrt{1-\frac{v^2}{c^2}}$ |
| 2. Schwarzschild | Infinity | $\Delta \tau(r)=100\sqrt{1-\frac{2GM}{rc^2}}$ |
| 3. Lorentz / SR | Moving frame \(v_{\text{ref}}\) | $\Delta t(v\mid v_{\text{ref}})=\dfrac{100}{\sqrt{1-\frac{1}{c^2}\left(\frac{v_{\text{ref}}-v}{1-\frac{v_{\text{ref}}v}{c^2}}\right)^2}}$ |
| 4. Schwarzschild | Finite radius \(r_A\) | $\Delta \tau(r\mid r_A)=100\sqrt{\dfrac{1-\frac{2GM}{rc^2}}{1-\frac{2GM}{r_Ac^2}}}$ |

---

## 1. Lorentz / SR with the absolute inertial reference frame as the reference

Assume **100 s** elapsed in the absolute inertial reference frame.

For each moving observer,

\[
\rho = \sqrt{1-\frac{v^2}{c^2}}, \qquad
\Delta \tau = \rho \cdot 100, \qquad
\Delta t_{\text{abs}} = \frac{\Delta \tau}{\rho}.
\]

| Observer | \(v/c\) | \(\rho=\sqrt{1-v^2/c^2}\) | Local proper time \(\Delta \tau\) (s) | Recovered absolute-frame time (s) |
| --- | ---: | ---: | ---: | ---: |
| Absolute frame | 0.0 | 1.0000 | 100.00 | 100.00 |
| Observer B | 0.5 | 0.8660 | 86.60 | 100.00 |
| Observer C | 0.7 | 0.7141 | 71.41 | 100.00 |
| Observer D | 0.9 | 0.4359 | 43.59 | 100.00 |

**Observation:** all observers recover the same **100 s** relative to the absolute inertial reference frame.

---

## 2. Schwarzschild with infinity as the reference

Assume **100 s** elapsed at infinity.


| Observer | \(\alpha = \frac{2GM}{rc^2}\) | \(\rho=\sqrt{1-\alpha}\) | Local proper time \(\Delta \tau\) (s) | Recovered time at infinity (s) |
| --- | ---: | ---: | ---: | ---: |
| Infinity | 0.0 | 1.0000 | 100.00 | 100.00 |
| Radius \(r_A\) | 0.1 | 0.9487 | 94.87 | 100.00 |
| Radius \(r_B\) | 0.2 | 0.8944 | 89.44 | 100.00 |
| Radius \(r_C\) | 0.3 | 0.8367 | 83.67 | 100.00 |

**Observation:** all radii recover the same **100 s** relative to infinity.

---

## 3. Lorentz / SR with a moving frame (\(0.9c\)) chosen as the reference

Assume **100 s** elapsed on the clock of the observer moving at \(0.9c\).


| Observer | Own speed | Relative speed to \(0.9c\) | \(\gamma_{\text{rel}}\) | Time assigned in that observer’s frame to 100 s on the \(0.9c\) clock |
| --- | ---: | ---: | ---: | ---: |
| \(0.5c\) observer | 0.5c | 0.7273c | 1.4552 | 145.52 s |
| \(0.6c\) observer | 0.6c | 0.6522c | 1.3188 | 131.88 s |
| \(0.7c\) observer | 0.7c | 0.5405c | 1.1882 | 118.82 s |
| \(0.9c\) observer | 0.9c | 0 | 1.0000 | 100.00 s |

**Observation:** the other moving observers do **not** agree with each other on the elapsed time relative to the \(0.9c\) reference clock.

---

## 4. Schwarzschild with a finite radius \(r_A\) chosen as the reference

Assume **100 s** elapsed at the reference radius \(r_A\).


| Pairwise comparison from \(r_A\) | \(\alpha_r\) | Factor \( \sqrt{\frac{1-\alpha_r}{1-\alpha_A}} \) | Elapsed time at that radius when \(r_A=100\) s |
| --- | ---: | ---: | ---: |
| \(r_A \to r_A\) | 0.1 | 1.0000 | 100.00 s |
| \(r_A \to \infty\) | 0.0 | 1.0541 | 105.41 s |
| \(r_A \to r_B\) | 0.2 | 0.9428 | 94.28 s |
| \(r_A \to r_C\) | 0.3 | 0.8819 | 88.19 s |

**Observation:** when \(100\) s elapsed at \(r_A\), the elapsed times at infinity, \(r_B\), and \(r_C\) are all different. So the pairwise treatment relative to a finite-radius reference does **not** produce a universal common elapsed time at the other radii.
