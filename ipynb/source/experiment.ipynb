## **Section I ‚Äî Empirical Observation**

### **Proposition 1 (Anomalous Observation)**
In repeated observations of vehicles within a defined region, the measured frequency of cars displaying a specific number-plate pattern significantly exceeded the predicted probability derived from the binomial distribution.

---

### **Proposition 2 (Statistical Expectation)**

Let the probability \( p \) of observing a vehicle with a particular number-plate pattern be \( p = 0.01 \),  
and let \( n = 1000 \) represent the number of observed vehicles.  
According to the binomial theorem:

$$
E[X] = n p = 10, \qquad 
\sigma = \sqrt{n p (1 - p)} \approx 3.15
$$

Hence, the number of occurrences \( X \) should, with overwhelming probability, lie between \( 4 \) and \( 16 \).

---

### **Proposition 3 (Observed Deviation)**

The actual count of vehicles matching the pattern exceeded \( 200 \) within \( n = 1000 \) observations,  
representing a deviation of more than \( 60 \) standard deviations from the expected mean.  
Such an occurrence lies beyond any reasonable statistical likelihood in the classical or relativistic physical sense.

## üî¢ Cross-Entropy and Entropic Bias in Physical Systems

We start from the expected (true) probability of observing a rare outcome (e.g., a unique number-plate type):

$$
P = [0.995, \, 0.005]
$$

and an **observed** (empirical) probability distribution, where the rare event becomes more frequent:

$$
Q = [0.7, \, 0.3]
$$

### üìê Cross-Entropy Calculation

The cross-entropy \( H(P, Q) \) is given by:

$$
H(P, Q) = - \sum_i P_i \log_2(Q_i)
$$

Substituting:

$$
H(P, Q) = - \big[ (0.995)\log_2(0.7) + (0.005)\log_2(0.3) \big]
$$

Compute each term:

$$
(0.995)\log_2(0.7) = 0.995 \times (-0.5146) = -0.5120
$$

$$
(0.005)\log_2(0.3) = 0.005 \times (-1.737) = -0.008685
$$

Now sum and negate:

$$
H(P, Q) = -(-0.5120 - 0.008685)
$$

$$
\boxed{H(P, Q) = 0.5207 \text{ bits}}
$$

### üîç Interpretation

In **information theory**, this means that encoding events from a true process with \( P=[0.995, 0.005] \) but assuming \( Q=[0.7, 0.3] \) requires an average of **0.52 bits per event**, rather than the optimal **0.045 bits** if \( Q=P \).  
This inefficiency represents a **divergence between expectation and reality** ‚Äî a misalignment of informational models.

In **physical terms**, however, if the rare outcome‚Äôs observed frequency rises from \(0.005\) to \(0.3\), the system becomes **more concentrated** around that event.  
Although Shannon entropy rises (more uncertainty about which event you‚Äôll see), **configurational entropy** decreases because spatial possibilities collapse toward a single dominant configuration.

Thus:

$$
\text{Information Entropy} \uparrow \quad \Longleftrightarrow \quad \text{Physical Entropy} \downarrow
$$

This duality underlies the idea that **information can bias physical order**.

## ‚öôÔ∏è Binomial Verification via Simulation

Below is a Python simulation of the same principle, showing how a rare-event probability distribution behaves under repeated trials.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n_experiments = 10000
n_flips = 1000
p = 0.005  # 0.5% chance for rare side

# Simulate binomial outcomes
rand_distB = np.random.binomial(n_flips, p, n_experiments)

# Compute exact binomial probabilities for k = 0..max observed
max_k = max(rand_distB)
k_values = np.arange(0, max_k + 1)
binom_probs = binom.pmf(k_values, n_flips, p)

# Print exact probabilities to 15 decimal places
print("k\tProbability")
for k, prob in zip(k_values, binom_probs):
    print(f"{k}\t{prob:.15f}")

# Plot histogram normalized to density
counts, bins, patches = plt.hist(
    rand_distB,
    bins=range(0, max_k + 2),
    color='skyblue',
    edgecolor='black',
    alpha=0.7,
    density=True
)

# Convert y-axis from fraction to percentage
ax = plt.gca()
yticks = ax.get_yticks()
ax.set_yticks(yticks)
ax.set_yticklabels([f'{y*100:.0f}' for y in yticks])

plt.ylabel('Percentage of experiments (%)')
plt.xlabel('Number of times coin lands on rare side')
plt.title('Simulated Binomial Distribution (n=100, p=0.005)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
