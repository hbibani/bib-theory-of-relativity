The persistence of incompatible local distributions therefore motivates **non-local informational coupling**, where external information (not intrinsic mechanics alone) modulates realized outcomes.

Finally, the document connects the empirical deviation to an **entropic/informational bias** formalized via reweighting (e.g., cross-entropy divergence between expected (P) and observed (Q)), interpreting the anomaly as a computationally induced concentration of realized configurations. The subsequent propositions formalize these claims and outline simulation-based verification against binomial expectation.
