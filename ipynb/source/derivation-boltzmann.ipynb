


## Derivation

Start with the Boltzmann-form probability (up to normalization):

$$
p_{\text{exp}} \propto e^{-E_{\text{exp}}/(k_B T)}.
$$

Assume the observed probability is dilated by a factor $\gamma$:

$$
p_{\text{obs}} = \gamma\, p_{\text{exp}}.
$$

Substitute the Boltzmann form into the dilation:

$$
p_{\text{obs}} \propto \gamma\, e^{-E_{\text{exp}}/(k_B T)}.
$$

Rewrite the multiplicative factor $\gamma$ as an exponential:

$$
\gamma = e^{\ln \gamma}.
$$

Combine exponentials:

$$
p_{\text{obs}} \propto e^{\ln\gamma}\, e^{-E_{\text{exp}}/(k_B T)}
= e^{\ln\gamma - E_{\text{exp}}/(k_B T)}.
$$

Now express $p_{\text{obs}}$ in canonical Boltzmann form with an *effective* energy $E_{\text{eff}}$:

$$
p_{\text{obs}} \propto e^{-E_{\text{eff}}/(k_B T)}.
$$

Match the exponents:

$$
\ln\gamma - \frac{E_{\text{exp}}}{k_B T} = -\frac{E_{\text{eff}}}{k_B T}.
$$

Multiply both sides by $-k_B T$:

$$
-k_B T \ln\gamma + E_{\text{exp}} = E_{\text{eff}}.
$$

Define the energy shift $\Delta E$ by $E_{\text{eff}} = E_{\text{exp}} + \Delta E$. Then:

$$
E_{\text{exp}} + \Delta E = E_{\text{exp}} - k_B T \ln\gamma
\Rightarrow\quad \boxed{\Delta E} = \boxed{-k_B T \ln\gamma}.
$$

### Interpretation

- If $\gamma>1$, then $\ln\gamma>0\Rightarrow \Delta E<0$: the state is effectively *lowered* in energy (favored).
- If $0<\gamma<1$, then $\Delta E>0$: the state is effectively *raised* in energy (suppressed).
