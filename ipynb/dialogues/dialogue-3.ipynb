# Title: On the Limited Access of Low-Energy Fields to Macroscopic Information, Their Need for Organisms to Acquire It, and On the Physical Origin of Laws

Note: By macroscopic is meant larger bodies, and by microscopic is meant smaller bodies.

------------------------------------------------------

Proposition 1: A moral law is the course of action most appropriate or fitting in a given case.

Proposition 2: A sentiment is the psychological impression produced in a person by an action.

 -Violation: the pain, discomfort, or uneasiness a person experiences in response to the conduct of someone or something else.

-Approval: the approbation, happiness, or pleasure a person experiences in response to that conduct.

------------------------------------------------------

Kaspur: A person’s morals are guided by their sentiments, rather than by reason.

Kali: That, indeed, was the ground of my suggestion: that the diversity of opinion in such matters arises from a diversity of sentiment. Thus, one person may entertain a more liberal judgment concerning what others may wear, while another, by reason of the uneasiness or offence he experiences at the sight of such liberty, may adopt a more severe and restrictive view. From this it would seem to follow that such judgments are determined by human beings themselves, and not by anything external to them.

Kaspur: That conclusion does not follow. Human beings are but the vehicles by which laws are discovered; and we commonly observe that the clearer and more perfect laws are delivered by those who are esteemed to possess the qualities necessary for right judgement and sound direction. The sentiments are but constructions of the field, designed to supply direction to a being whose understanding is attended with many defects...it was they — the fields — that fashioned and attended to mankind, and which, through the use of his senses, led him to contrive those devices by which the body is protected with garments...It is their laws, not human constructs; and by understanding that, you will understand how to alleviate your confusion.

Kali: Very well. But, speaking now of vehicles, I was instructed to inquire of you concerning the fields themselves, and particularly concerning the limits of their operation.

Kaspur: Yes, I did wish to show you something of that matter. You know already that biological organisms had a simpler origin. And if we look at this

1. Low-energy fields would most plausibly have begun by constructing organisms at very small scales.
2. As organisms developed, a substantial portion of the operational computation became offloaded onto the organism itself, thereby increasing its effective computational power.
3. It is precisely this offloaded computational component within biological organisms that makes stable macroscopic existence possible.

Kali: Yes, certainly.

Kaspur: The domain of the field is exceedingly narrow; and the macroscopic organisms which now appear before us are the consequence of a careful and gradual interaction of these fields and the bodies which they have constructed, from their first commencement down to the forms we presently observe. Its vision also is impaired...

Kali: By what means can its vision be said to be impaired?...how can a field with such power and intelligence be limited in its vision? 

Kaspur: Oh...you didn't know?...they are distinct from god...they require these larger bodies to gain a feel for the macroscopic world..for whatever macroscopic information it must acquire from the world, it cannot obtain immediately by itself, but must receive such information through the senses of the organisms it has constructed; and for that reason, considerable resources are required in order to secure the advantages, or to avoid the dangers, which present themselves with respect to its continued existence; and it effects this by directing their movements, and by making use of the lessons contained in their history, in order to determine what plans must be laid for its own preservation.
