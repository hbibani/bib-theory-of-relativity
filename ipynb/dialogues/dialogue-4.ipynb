## Title: On Religious Authority

---

## Please refer to psychology-case-godel in source folder

---

Proposition 1: Authority may be understood as that quality in a person by which he is fitted to direct, instruct, or guide others in some particular manner.

Proposition 2: Religious authority may be understood as that quality in a person by which he is fitted to instruct others concerning the relation between the world, the creatures contained in it, and the meaning that is supposed to arise from both.

---

Kaspur: And by what mark do you judge a doctrine to be worthy of your attention?

Kali: I have confidence only in that testimony which arises from the immediate impressions of my senses.

Kaspur: And upon what grounds may a thing be refuted?

Kali: When that which is represented stands in contradiction to the testimony furnished by my own experience.

Kaspur: And what say you to those principles such as the physical laws, though not immediately perceived, are yet derived from the world itself?

Kali: I admit those likewise, as constituting a part of my testimony.

Kaspur: And from your own account, it follows that the physical laws constitute a part of your testimony?

Kali: It forms the major part of it.

Kaspur: And you allow, then, that there are certain principles which remain yet to be discovered?

Kali: I do.

Kaspur: And do you believe that the most important of all principles may still lie undiscovered?

Kali: That is my opinion.

Kaspur: And if a principle, perhaps the most important of all, still lies undiscovered, and in the absence of that principle, you might receive as true what is in reality false. And so upon what foundation can your testimony ever be trusted? 

Kali: My belief extends no farther than to that which agrees with what I understand to be true.

Kaspur: But here we are presented with a case, wherein, by a careful derivation, some opinion is brought into dispute which you had formerly received as physically true. And of what use is that information, if the principles in question can arise only from a diligent examination of that case; so that, though it may suffice to correct your former apprehension, it yet leaves you incapable of forming those more difficult principles from which it is derived?

Kali: I don't understand.

Kaspur: If it were laid down as a rule, that any person who claims religious authority """"must be exempt from death or destruction"""", from when, in the case presented to you, could there be found any just agreement with the testimony of your observation? [REFER PROPOSITION 2]

Kali: I had seen no more than the strange motion of the cars; from what source, then, could such a principle as exemption from death be derived...only by an extension from that strange motion of cars?

Kaspur: And were those from whom you sought instruction concerning the world, the beings contained in it, and the relation between the two, appeared to be exempt from death?** [REFER PROPOSITION 2]

Kali: No...but that is not a principle.
