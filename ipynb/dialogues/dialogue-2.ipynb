## Title: Theology in the analysis of parasitic organisms 

---------------------------------------------

Proposition 1: Knowledge is the ability to recognise and understand the composition of something.

Corollary 1: There are two types of knowledge (see Demonstration):

1. Direct / Conceptual: when the thing being understood is examined directly in terms of its own nature, and not through the consequences of something which produces it.
2. Relational: when the thing being understood is examined through other things which produce or affect it, and which are distinct from the thing itself.

Demonstration 1:

A person’s character may be understood distinctly from the thing which causes it (for example, the body). In this sense, character can be considered a conceptual object of knowledge rather than a relational one. However, a human character is also relational because it is composed of things such as the brain and the limbic system. This means that character as an idea or concept is different from human character as it exists in the physical organism.

The study of the body is therefore a relational type of knowledge, because the study of its components necessarily involves relations to the components of other things. For example, the brain requires oxygen to perform its functions. When we study the components of the brain—which are the source of a person’s character—we must also consider oxygen in the atmosphere, which is something distinct from the brain itself. In this way, the knowledge involved is relational.

AGENTS

Proposition 6: An agent is anything that can be viewed as perceiving its environment through sensors and acting upon that environment through actuators.

Corollary 1: There are two types of agents:

1. Originate / Central — an agent that has the ability to understand biological components and construct biological organisms.
2. Peripheral — an agent whose character extends from the activity of an organ (for example, human character).

Corollary 2: Accessibility is the degree, scope, or limitation within which information can be obtained.

Corollary 3: An originate or central agent understands the knowledge of a peripheral agent.

Demonstration 1: A person (a peripheral agent) is familiar with the term Netflix as a platform through which movies can be accessed. The term has the same meaning to a central agent. The central agent is the being involved in producing the person.

Proposition 7: A precept sequence is the complete history of information that originates from a sense.

Proposition 8: A central agent is omniscient—that is, it is aware of the actual outcomes of its actions [REFER Proposition 1, Corollary 1].

Proposition 9: Information gathering is the process of using acquired information in order to form a decision.

Proposition 10: Learning is the process of acquiring knowledge and improving performance through reflection on what has been perceived [REFER Proposition 7].

KNOWABILITY IN THE ENVIRONMENT

Proposition 11: The exact nature of the limitations in the visibility of the environment is not known in full. There is speculative evidence suggesting that agents involved in the production of biological organisms are not omnipresent, and therefore do not have access to all information. Instead, they appear to experience only a subset of the total information that exists.
This implies that there would be limitations in their ability to produce organs, due to a lack of complete information. There are also indications that such agents possess a finite amount of resources. This is evident from the fact that all organisms are finite in power, which suggests that the producer may possess a similar limitation.

MORAL CHARACTER

Proposition 12: An agent’s moral character is determined by the intention of its objectives, and moral character can be identified by the way it uses the information it possesses [REFER Proposition 5].

Scholium 1: It is not easy to determine the intentions of a divine agent, because there is no direct perception of its thoughts. Therefore, the intentions of a central or originate agent must be inferred from the things it constructs, since these constructions provide insight into how the information it has gathered was used [REFER Proposition 12 and Proposition 7].

Demonstration 1: The producer of the ichneumon wasp utilized information gathered from its senses and determined that the butterfly larvae smelled the same as the ant larvae. It consumed both the butterfly and the ant, and from this information understood the mechanical weaknesses of their bodies. Consequently, it formulated a plan to construct an organ that paralyzes the ant’s movement, as well as a mechanism that allows its egg to thrive within the butterfly larvae.

One of the principal outcomes of carrying out such a plan is the parasitic destruction of another organism. Therefore, the manner in which the information was used, after it had been gathered (about the smells of the two larvae), can be considered evil [REFER Proposition 8 — regarding outcomes].

Proposition 13: There is a clear difference in the moral character of agents involved in the creation of biological organisms (central or originate agents).

Scholium 1: There may be multiple administrations involved in producing organisms, each possessing different moral characters. This implies that there could be different administrations responsible for producing biological organisms, including ones whose actions could be considered evil.
