# Set Theory in the Bibs-Processing Framework

* **Proposition 1:**  We define a finite state set, as a sub frame, which is a fundemental unit of an entire frame:

$$
S_{subframe}

Each state $S_{subframe}$ has two intrinsic parameters:
$$
\begin{array}{ll}

* **Corollary 1:**  We can calculate bibs to proper time by the following

Each computational tick that executes state $S_i$ consumes $b_i$ absolute bibs and advances proper time by:

$$
\Delta \tau_i = s_i \, b_i
$$

Conversely, if we measure elapsed proper time and know bibs consumed, the processing factor can be recovered:

$$
s_i = \frac{\Delta \tau_i}{b_i}
$$


* **Proposition 2:**  The system evolves through a discrete sequence of frames, a total frame is defined by subframes within an entire frame, and each frame takes a fixed amount of bibs to complete, therefore each state sets its bib-time to the same value of that frame:

$$
F = (S_1, S_2, \dots, S_n)

 Where the total Frames in a given sequence is given by:

$$

$$


* **Corollary 1:**  Each corresponds to one stateâ€™s computation. The per-state tick is:

$$
\Delta \tau_i = s_i \, b_i

After $k$ steps, cumulative sums are, where the total amount of bibs is the same within that frame:

$$
B = b, \qquad 
\Tau_k = \sum_{i=1}^{k} \Delta \tau_i = \sum_{i=1}^{k} s_i \, b_i
$$

The average processing-speed factor over a given frame is:

$$
\bar{s}_{(1..k)} = \frac{\Tau_k}{B}

```python
# Example A â€” Global Scaling Factor
import numpy as np

# Example A
b_A = np.array([10, 10, 10, 10])
s_A = 0.8

# Per-state proper times
delta_tau_A = s_A * b_A

# Cumulative sums
B_cumsum_A = np.cumsum(b_A)
Tau_cumsum_A = np.cumsum(delta_tau_A)
s_bar_A = Tau_cumsum_A / B_cumsum_A

print("Example A â€” Global Scaling Factor")
print("Per-state proper times:", delta_tau_A)
for i in range(len(b_A)):
    print(f"B_{i+1} = {B_cumsum_A[i]}, Tau_{i+1} = {Tau_cumsum_A[i]}, s_bar_{i+1} = {s_bar_A[i]}")
print("Invariant check: Tau_4 / B_4 =", Tau_cumsum_A[-1]/B_cumsum_A[-1], "\n")

# Example B â€” Per-State Scaling Factors
b_B = np.array([10, 10, 10, 10])
s_B = np.array([0.9, 0.8, 0.6, 0.8])

# Per-state proper times
delta_tau_B = s_B * b_B

# Cumulative sums
B_cumsum_B = np.cumsum(b_B)
Tau_cumsum_B = np.cumsum(delta_tau_B)
s_bar_B = Tau_cumsum_B / B_cumsum_B

print("Example B â€” Per-State Scaling Factors")
print("Per-state proper times:", delta_tau_B)
for i in range(len(b_B)):
    print(f"B_{i+1} = {B_cumsum_B[i]}, Tau_{i+1} = {Tau_cumsum_B[i]}, s_bar_{i+1} = {s_bar_B[i]}")
print("Invariant check: Tau_4 / B_4 =", Tau_cumsum_B[-1]/B_cumsum_B[-1])
```

## ðŸ”¢ Cross-Entropy and Entropic Bias in Physical Systems

We start from the expected (true) probability of observing a rare outcome (e.g., a unique number-plate type):

$$
P = [0.995, \, 0.005]
$$

and an **observed** (empirical) probability distribution, where the rare event becomes more frequent:

$$
Q = [0.7, \, 0.3]
$$

### ðŸ“ Cross-Entropy Calculation

The cross-entropy \( H(P, Q) \) is given by:

$$
H(P, Q) = - \sum_i P_i \log_2(Q_i)
$$

Substituting:

$$
H(P, Q) = - \big[ (0.995)\log_2(0.7) + (0.005)\log_2(0.3) \big]
$$

Compute each term:

$$
(0.995)\log_2(0.7) = 0.995 \times (-0.5146) = -0.5120
$$

$$
(0.005)\log_2(0.3) = 0.005 \times (-1.737) = -0.008685
$$

Now sum and negate:

$$
H(P, Q) = -(-0.5120 - 0.008685)
$$

$$
\boxed{H(P, Q) = 0.5207 \text{ bits}}
$$

### ðŸ” Interpretation

In **information theory**, this means that encoding events from a true process with \( P=[0.995, 0.005] \) but assuming \( Q=[0.7, 0.3] \) requires an average of **0.52 bits per event**, rather than the optimal **0.045 bits** if \( Q=P \).  
This inefficiency represents a **divergence between expectation and reality** â€” a misalignment of informational models.

In **physical terms**, however, if the rare outcomeâ€™s observed frequency rises from \(0.005\) to \(0.3\), the system becomes **more concentrated** around that event.  
Although Shannon entropy rises (more uncertainty about which event youâ€™ll see), **configurational entropy** decreases because spatial possibilities collapse toward a single dominant configuration.

This duality underlies the idea that **information can bias physical order**.

## âš™ï¸ Binomial Verification via Simulation

Below is a Python simulation of the same principle, showing how a rare-event probability distribution behaves under repeated trials.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n_experiments = 10000
n_flips = 1000
p = 0.005  # 0.5% chance for rare side

# Simulate binomial outcomes
rand_distB = np.random.binomial(n_flips, p, n_experiments)

# Compute exact binomial probabilities for k = 0..max observed
max_k = max(rand_distB)
k_values = np.arange(0, max_k + 1)
binom_probs = binom.pmf(k_values, n_flips, p)

# Print exact probabilities to 15 decimal places
print("k\tProbability")
for k, prob in zip(k_values, binom_probs):
    print(f"{k}\t{prob:.15f}")

# Plot histogram normalized to density
counts, bins, patches = plt.hist(
    rand_distB,
    bins=range(0, max_k + 2),
    color='skyblue',
    edgecolor='black',
    alpha=0.7,
    density=True
)

# Convert y-axis from fraction to percentage
ax = plt.gca()
yticks = ax.get_yticks()
ax.set_yticks(yticks)
ax.set_yticklabels([f'{y*100:.0f}' for y in yticks])

plt.ylabel('Percentage of experiments (%)')
plt.xlabel('Number of times coin lands on rare side')
plt.title('Simulated Binomial Distribution (n=100, p=0.005)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
