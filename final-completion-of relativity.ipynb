### Reference Frames and Time

* **Definition 1 (Frame):**  
A frame is the smallest complete unit of computation within a reference frame.
Each frame represents a discrete ‚Äúrendering‚Äù of physical reality, analogous to a frame in digital video, and requires a certain number of bibs to be computed.

* **Definition 2 (Tick):**  
A tick is the completion of one frame.
It marks the advancement of proper time in that reference frame. The length of a tick depends on how many bibs the frame consumes.

* **Proposition 2 (Reference Frames as Processes):**  
  A physical reference frame corresponds to a process that consumes a certain number of bibs per tick.

* **Proposition 3:**  
  Two reference frames which require different computational times will experience different tick rates, producing relative time dilation.

* **Proposition 4 (Frame Rendering Analogy):**  
  Each reference frame consists of sequential ‚Äúframes‚Äù that must be computed.  
  The duration of a tick depends on the number of bibs required to complete one frame.

* **Corollary 1:**  
  If more bibs are required to complete a process, ticks occur more slowly.

# üîÄ Bibs-Based Relativity Experiment ‚Äî Frames, Ticks, and Mathematical Consistency

This experiment illustrates the **bibs-processing model**.

* A **stationary observer** measures 50 seconds of proper time.
* A **train observer** is moving with processing ratio (p = 0.9).
* The scaling factor is in relation to velocity only, where we ignore other processing requirements of, energy, life and matter setting them to zero.

$$
s(v) = \sqrt{1-\frac{v^2}{u^2}}
$$

* The moving observer‚Äôs proper time is:

$$
\tau_{\text{train}} = \tau_{\text{stationary}} \cdot s(v)
$$

* The **absolute bibs** are computed as:

$$
\Delta b = \frac{\Delta \tau}{s(v)}
$$

---

### Concepts in Terms of Frames and Ticks


* **(Absolute Reference Frame):**  
  There exists an underlying computational space where all processes are executed.  
  This space operates in **absolute time**, discretized into fundamental units called **bibs**.

* A **frame** is one computational rendering of reality in a given reference frame.
* A **tick** is the completion of a frame; it marks one advance of proper time for that observer.
* Each frame requires some number of bibs to complete. If a frame consumes more bibs, ticks occur more slowly.

---

### Interpretation of the Experiment

Both observers traverse the **same number of absolute bibs**, even though their experienced proper times differ.

* The **stationary observer** (50 s) produced more ticks (more frames completed within that span).
* The **train observer** (15.81 s) produced fewer ticks (fewer frames completed within the same absolute bibs).

This shows:

* Bibs are conserved across observers (absolute resource).
* Time dilation arises because moving frames require more bibs per tick, slowing their tick rate.
* The stationary frame renders more frames within the same bib count, while the moving frame renders fewer.

---




```python
import numpy as np

# Parameters
p_train = 0.9   # directly use p = 0.9
s_train = np.sqrt(1 - p_train)
s_stationary = 1.0

# Stationary observer measures 50 seconds
tau_stationary = 50

# Train observer proper time
tau_train = tau_stationary * s_train

# Convert both to bibs using Œît = ŒîœÑ / s(v)
bibs_stationary = tau_stationary / s_stationary
bibs_train = tau_train / s_train

print(f"Scaling factor (s) for train: {s_train:.3f}")
print(f"Stationary observer time: {tau_stationary:.2f} s")
print(f"Train observer time: {tau_train:.2f} s")
print(f"Stationary bibs: {bibs_stationary:.1f}")
print(f"Train bibs: {bibs_train:.1f}")
```

# Set Theory in the Bibs-Processing Framework

* **Proposition 1:**  We define a finite state set, as a sub frame, which is a fundemental unit of an entire frame:

$$
S_{subframe}
$$

Each state $S_{subframe}$ has two intrinsic parameters:
$$
\begin{array}{ll}
b_i \in \mathbb{N}_{>0}, & \text{number of bibs required to compute that state} \\
s_i \in (0,1], & \text{processing-speed factor (local clock rate or time dilation)}
\end{array}
$$

* **Corollary 1:**  We can calculate bibs to proper time by the following

Each computational tick that executes state $S_i$ consumes $b_i$ absolute bibs and advances proper time by:

$$
\Delta \tau_i = s_i \, b_i
$$

Conversely, if we measure elapsed proper time and know bibs consumed, the processing factor can be recovered:

$$
s_i = \frac{\Delta \tau_i}{b_i}
$$


* **Proposition 2:**  The system evolves through a discrete sequence of frames, a total frame is defined by subframes within an entire frame, and each frame takes a fixed amount of bibs to complete, therefore each state sets its bib-time to the same value of that frame:

$$
F = (S_1, S_2, \dots, S_n)
$$

 Where the total Frames in a given sequence is given by:

$$
F_{total} = (F_1, F_2, \dots, F_n)
$$


* **Corollary 1:**  Each corresponds to one state‚Äôs computation. The per-state tick is:

$$
\Delta \tau_i = s_i \, b_i
$$

After $k$ steps, cumulative sums are, where the total amount of bibs is the same within that frame:

$$
B = b, \qquad 
\Tau_k = \sum_{i=1}^{k} \Delta \tau_i = \sum_{i=1}^{k} s_i \, b_i
$$

The average processing-speed factor over a given frame is:

$$
\bar{s}_{(1..k)} = \frac{\Tau_k}{B}
$$


```python
# Example A ‚Äî Global Scaling Factor
import numpy as np

# Example A
b_A = np.array([10, 10, 10, 10])
s_A = 0.8

# Per-state proper times
delta_tau_A = s_A * b_A

# Cumulative sums
B_cumsum_A = np.cumsum(b_A)
Tau_cumsum_A = np.cumsum(delta_tau_A)
s_bar_A = Tau_cumsum_A / B_cumsum_A

print("Example A ‚Äî Global Scaling Factor")
print("Per-state proper times:", delta_tau_A)
for i in range(len(b_A)):
    print(f"B_{i+1} = {B_cumsum_A[i]}, Tau_{i+1} = {Tau_cumsum_A[i]}, s_bar_{i+1} = {s_bar_A[i]}")
print("Invariant check: Tau_4 / B_4 =", Tau_cumsum_A[-1]/B_cumsum_A[-1], "\n")

# Example B ‚Äî Per-State Scaling Factors
b_B = np.array([10, 10, 10, 10])
s_B = np.array([0.9, 0.8, 0.6, 0.8])

# Per-state proper times
delta_tau_B = s_B * b_B

# Cumulative sums
B_cumsum_B = np.cumsum(b_B)
Tau_cumsum_B = np.cumsum(delta_tau_B)
s_bar_B = Tau_cumsum_B / B_cumsum_B

print("Example B ‚Äî Per-State Scaling Factors")
print("Per-state proper times:", delta_tau_B)
for i in range(len(b_B)):
    print(f"B_{i+1} = {B_cumsum_B[i]}, Tau_{i+1} = {Tau_cumsum_B[i]}, s_bar_{i+1} = {s_bar_B[i]}")
print("Invariant check: Tau_4 / B_4 =", Tau_cumsum_B[-1]/B_cumsum_B[-1])
```

# Example ‚Äî Effect of processing life has on a Subframe within a Frame

* **Proposition 3:**  
Let a frame \(F\) be composed of 5 subframes:  

$$
F = (S_1, S_2, S_3, S_4, S_5)
$$

Each subframe consumes an equal number of bibs:  

$$
B = 20
$$

---

* **Proposition 4:**  
Now, let subframe $S_3$ represent a subsystem within the frame that carries a local life-processing requirement:

$$
p_{\text{life}} = 0.01
$$

According to the global definition of processing-speed factor:

$$
s(v) = \sqrt{1 - p_{\text{life}}}
$$

Therefore, subframe $S_3$ experiences a local slowdown:

$$
s_3 = \sqrt{1 - 0.01} = 0.994987
$$

All other subframes remain at full processing speed $s_i = 1.0$.  

---

* **Corollary 2:**  
The per-state proper time is given by:

$$
\Delta \tau_i = s_i \, b_i
$$

Hence, the total proper time for this frame becomes:

$$
\begin{aligned}
\Delta \Tau_{\text{case}} 
 &= (0.994987 \times 20) 
&= 19.89974
\end{aligned}
$$


* **Interpretation:**  
Even a small local life-processing fraction $p_{\text{life}} = 0.01$ within a single subframe reduces the total proper time per frame while maintaining a constant bib budget $B$.  

The number of frames per absolute-bib flow remains unchanged, but the proper-time advancement within each frame decreases.  
This represents a localized computational slowdown or **time dilation**, emerging from life-processing demand within the subframe $S_3$.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Frame setup ===
subframes = ['S1','S2','S3 (Si)','S4','S5']
B = 100.0
b = np.array([20.0]*5)  # equal bib allocation per subframe

# Baseline (all at s=1)
s_baseline = np.array([1.0]*5)
delta_tau_baseline = s_baseline * b
Tau_baseline = delta_tau_baseline.sum()

# Case: S3 has p_life = 0.01
p_life = 0.01
s_s3 = np.sqrt(max(0.0, 1.0 - p_life))
s_case = s_baseline.copy()
s_case[2] = s_s3
delta_tau_case = s_case * b
Tau_case = delta_tau_case.sum()

# === Results ===
df = pd.DataFrame({
    "Subframe": subframes,
    "b (bibs)": b,
    "s (baseline)": s_baseline,
    "ŒîœÑ (baseline)": delta_tau_baseline,
    "s (p_life=0.01 on S3)": s_case,
    "ŒîœÑ (case)": delta_tau_case
})

# display_dataframe_to_user("Simple example: S3 with p_life = 0.01", df)

diff = Tau_baseline - Tau_case
pct_drop = diff / Tau_baseline * 100.0

print(f"Total proper time per frame (baseline): {Tau_baseline:.6f}")
print(f"Total proper time per frame (S3 p_life=0.01): {Tau_case:.6f}")
print(f"Absolute decrease per frame: {diff:.6f} ({pct_drop:.6f} % decrease)")

# === Visualization ===
indices = np.arange(len(subframes))
width = 0.35
plt.figure(figsize=(7,3.5))
plt.bar(indices - width/2, delta_tau_baseline, width, label='ŒîœÑ baseline')
plt.bar(indices + width/2, delta_tau_case, width, label='ŒîœÑ (S3 p_life=0.01)')
plt.xticks(indices, subframes)
plt.ylabel('ŒîœÑ (proper time advance)')
plt.title('Effect of p_life = 0.01 on S3 (ŒîœÑ per subframe)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Relativistic Time Dilation Example for 10 Seconds Lab Time

---

### Energy \& Lorentz Factor: 
Relativistic kinetic energy (energy on top of rest energy) is

$$
E_{\text{kin}} = (\gamma - 1) m c^2,
$$

which can be inverted to give $\gamma$ if the added energy is known:

$$
\gamma = 1 + \frac{E_{\text{kin}}}{m c^2}.
$$

---
* **Proposition 1:**  
Let a system of 1~mol of water be acted on by a minimal work $W_\text{min} = 6000\,\mathrm{J}$ to order it. The rest mass is

$$
m_\text{mol} = 0.018 \,\mathrm{kg}, \qquad c = 3 \times 10^8 \,\mathrm{m/s}.
$$

* **Corollary 1:** 
Lorentz factor:

$$
\gamma = 1 + \frac{W_\text{min}}{m_\text{mol} c^2} \approx 1.0000000000037.
$$

* **Proposition 2:**  
Proper time of water molecules relative to lab time $t_\text{lab}$ is

$$
\tau_\text{water} = \frac{t_\text{lab}}{\gamma}.
$$

* **Demonstration 1:**  
For $t_\text{lab} = 10\,\mathrm{s}$:

$$
\tau_\text{water} \approx \frac{10}{1.0000000000037} \approx 9.999999999963 \,\mathrm{s}.
$$

* **Corollary 1:** 
Elapsed time difference:

$$
\Delta t = t_\text{lab} - \tau_\text{water} \approx 10 - 9.999999999963 \approx 3.7 \times 10^{-11} \,\mathrm{s}.
$$

This demonstrates that even after 10~seconds of lab time, the relativistic time dilation due to ordering 1~mol of water is **extremely tiny**, only tens of picoseconds.
